# Image to Pencil sketch

## Import Libraries 

```python
import cv2
from matplotlib import pyplot as plt
```

## Reading the Image 

```python
image=cv2.imread("C:\\Users\K Yashwanth Reddy\OneDrive - Chandigarh University\Desktop\LION.jpg")
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(16,16))
plt.imshow(image)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

## Converting Original Image to Gray Scale

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(16,16))
plt.imshow(gray_image,cmap="gray")
plt.axis("off")
plt.title("Gray Image")
plt.show()
```

## Converting the Gray Scale Image to Negative Image

```python
invert_image =cv2.bitwise_not(gray_image)
plt.figure(figsize=(16,16))
plt.imshow(invert_image,cmap="gray")
plt.title("Inverted Image")
plt.axis("off")
plt.show()
```

## Converting Negative Scale Image to Blur Image

```python
blur_image=cv2.blur(invert_image,(21,21),0)
plt.figure(figsize=(16,16))
plt.imshow(blur_image,cmap="gray")
plt.axis("off")
plt.title("BlurImage")
plt.show()
```

## Converting Blur Image to Inverted Blur Image

```python
inverted_blur_image = cv2.bitwise_not(blur_image)
plt.figure(figsize=(16,16))
plt.imshow(inverted_blur_image,cmap="gray")
plt.axis("off")
plt.title("Inverted Blur Image")
plt.show()
```

## Creating the pencil sketch by mixing the grayscale image with the inverted blurry image 

```python
sketch=cv2.divide(gray_image,inverted_blur_image,scale=256.0)
plt.figure(figsize=(16,16))
plt.imshow(sketch,cmap="gray")
plt.title("Pencil Sketch Image")
plt.axis("off")
plt.show()

```

# Original Image vs Pencil sketch Image

```python


plt.figure(figsize=(8,8))

plt.imshow(image)
plt.axis("off")
plt.title("Original Image")
plt.show()
plt.figure(figsize=(8,8))
plt.imshow(sketch,cmap="gray")

plt.title("Pencil sketch Image")
plt.axis("off")
plt.show()
```
